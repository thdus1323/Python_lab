```python
import win32com.client
import pandas as pd

# 한글 열기
hwp = win32com.client.Dispatch("HWPFrame.HwpObject")
hwp.XHwpWindows.Item(0).Visible = True  # 한글 창을 화면에 표시
hwp.RegisterModule("FilePathCheckDLL", "FilePathCheckerModuleExample")

file_path = r"C:\workspace\한글자동화\python_lab\07. 한글 문서\❌누름틀 전_문서\[서식33] 인턴형 일경험 참여신청서_연습용으로 만듦.hwp"
hwp.Open(file_path, "HWP", "forceopen:true")

def 엑셀데이터불러오기():
    file_path_excel = r"C:\workspace\한글자동화\python_lab\07. 한글 문서\한글 자동화_엑셀파일_lsy.xlsm"  # 엑셀 파일 경로 지정
    sheet_name = "33. 인턴형 일경험 참여신청서"  # 엑셀 시트 이름
    header = 0
    excel_data = pd.read_excel(file_path_excel, sheet_name=sheet_name, header=header, dtype=str).fillna("")
    excel_data.columns = excel_data.columns.str.strip()
    return excel_data

# 엑셀 데이터를 불러옵니다.
excel_data = 엑셀데이터불러오기()
print(excel_data)

# 누름틀 필드에 값을 집어넣는 함수
def 누름틀필드에_값_집어넣기(hwp, excel_data):
    try:
        # 엑셀 A열에 있는 필드명과 B열에 있는 값을 한글 문서에 삽입
        for index, row in excel_data.iterrows():
            field_name = row.iloc[1]  # A열: 필드명 (index 1)
            field_value = row.iloc[2]  # B열: 값 (index 2)
            if pd.notna(field_name) and pd.notna(field_value):
                hwp.PutFieldText(field_name, str(field_value))
        # 성공 메시지 출력
        print("필드에 값을 성공적으로 삽입했습니다.")
    except Exception as e:
        # 오류 메시지 출력
        print(f"필드에 값 삽입 중 오류 발생: {e}")

# 함수 호출
누름틀필드에_값_집어넣기(hwp, excel_data)
```

```python
# 엑셀 데이터 불러오기 함수
def 엑셀데이터불러오기():
    file_path_excel = r"C:\workspace\한글자동화\python_lab\07. 한글 문서\한글 자동화_엑셀파일_lsy.xlsm"  # 엑셀 파일 경로 지정
    sheet_name = "33. 인턴형 일경험 참여신청서"  # 엑셀 시트 이름
    header = 10
    excel_data = pd.read_excel(file_path_excel, sheet_name=sheet_name, header=header, dtype=str).fillna("")
    excel_data.columns = excel_data.columns.str.strip()
    return excel_data

# 엑셀 데이터를 불러옵니다.
excel_data = 엑셀데이터불러오기()
print(excel_data)

# 최종학력구분에 따라 데이터를 한글 문서에 삽입하는 로직
for index, row in excel_data.iterrows():
    try:
        # 최종학력구분에 따라 기호를 추가 (대학생이면 O 표시, 고등학생이면 X 표시 등)
        if row.get('최종학력구분') == '대학교_4년제':
            hwp.PutFieldText('대학교', '■')
            hwp.PutFieldText('대학교명', row.get('최종학교명', ''))
            hwp.PutFieldText('학과명', row.get('최종학과', ''))
            if row.get('최종학력상태') == '졸업':
                hwp.PutFieldText('대학교졸업', '■')
            elif row.get('최종학력상태') == '졸업예정자':
                hwp.PutFieldText('졸업예정자', '■')
            elif row.get('최종학력상태') == '휴학':
                hwp.PutFieldText('휴학', '■')
            elif row.get('최종학력상태') == '재학':
                if row.get('학년') in ['1', '2']:
                    hwp.PutFieldText('대학교재학(1~2학년)', '■')
                elif row.get('학년') in ['3', '4']:
                    hwp.PutFieldText('대학교재학(3~4학년)', '■')

        elif row.get('최종학력구분') == '대학교_2년제':
            hwp.PutFieldText('대학교', '■')
            hwp.PutFieldText('대학교명', row.get('최종학교명', ''))
            hwp.PutFieldText('학과명', row.get('최종학과', ''))
        elif row.get('최종학력구분') == '대학교_2/3년제':
            hwp.PutFieldText('대학교', '■')
            hwp.PutFieldText('대학교명', row.get('최종학교명', ''))
            hwp.PutFieldText('학과명', row.get('최종학과', ''))

        elif row.get('최종학력구분') == '고등학교':
            hwp.PutFieldText('고등학교', '■')
            hwp.PutFieldText('고등학교명', row.get('최종학교명', ''))
            if row.get('최종학력상태') == '졸업':
                hwp.PutFieldText('고등학교졸업', '■')
            elif row.get('최종학력상태') == '재학':
                if row.get('학년') in ['1', '2']:
                    hwp.PutFieldText('고등학교재학(1~2학년)', '■')
                elif row.get('학년') == '3':
                    hwp.PutFieldText('고등학교재학(3학년)', '■')
            elif row.get('최종학력상태') == '검정고시':
                hwp.PutFieldText('검정고시', '■')

    except Exception as e:
        print(f"오류 발생 (행 {index + 1}): {e}")

# 엑셀 반복이 종료되면 한글 종료
# hwp.Quit()
```
