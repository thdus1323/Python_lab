```python
import win32com.client
import pandas as pd

# 한글 열기
hwp = win32com.client.Dispatch("HWPFrame.HwpObject")
hwp.XHwpWindows.Item(0).Visible = True  # 한글 창을 화면에 표시
hwp.RegisterModule("FilePathCheckDLL", "FilePathCheckerModuleExample")

file_path = r"C:\workspace\한글자동화\python_lab\07. 한글 문서\❌누름틀 전_문서\세로누름틀 테스트중.hwp"
hwp.Open(file_path, "HWP", "forceopen:true")

def 엑셀열기():
    file_path_excel = r"C:\workspace\한글자동화\python_lab\07. 한글 문서\한글 자동화_엑셀파일_lsy.xlsm"  # 엑셀 파일 경로 지정
    sheet_name = "125. (공통직무교육기관) 사전직무교육 실시보고"  # 엑셀 시트 이름
    header = 20
    excel_data = pd.read_excel(file_path_excel, sheet_name=sheet_name, header=header, dtype=str).fillna("")
    return excel_data

# 엑셀 데이터를 불러옵니다.
excel_data = 엑셀열기()
print(excel_data)
```

```python
def 세로누름틀생성(hwp, 시작인덱스 , 필드이름, 개수):
    for i in range(개수):
        field_name = f"{필드이름}_{시작인덱스 + i}"
        # hwp.HAction.Execute("MoveDown", hwp.HParameterSet.HFindReplace.HSet)
        
        # 누름틀 필드 생성 (현재 커서 위치에 삽입)
        hwp.CreateField(필드이름, "", field_name)
        print(f"Inserted field: {field_name}")
        
        hwp.HAction.Execute("MoveDown", hwp.HParameterSet.HFindReplace.HSet)

# '성별' 필드를 현재 커서 위치에서 시작으로 반복적으로 누름틀을 생성하고 텍스트를 설정
# 세로누름틀생성(hwp,시작인덱스=21,필드이름="이름",개수=30)
```

```python
def 참여자_명단_삽입():
        try:
            # 엑셀 데이터를 반복하여 한글 문서의 누름틀에 필드명과 값을 삽입
            for index, row in 엑셀열기().iterrows():
                # 추가적인 필드에 데이터 삽입 (성명, 생년월일, 성별, 참여일, 연락처)
                hwp.PutFieldText(f"이름_{index}", str(row[1]))
                # hwp.PutFieldText(f"생년월일_{index}", str(row[1]))
                # hwp.PutFieldText(f"성별_{index}", str(row[2]))
                # hwp.PutFieldText(f"참여일_{index}", str(row[3]))
                # hwp.PutFieldText(f"연락처_{index}", str(row[4]))
                print(hwp.PutFieldText(f"이름_{index}", str(row[1])))
            # 성공 메시지 출력
            print("참여자 명단을 한글 누름틀에 성공적으로 삽입했습니다.")
        except Exception as e:
            # 오류 메시지 출력
            print(f"참여자 명단 삽입 중 오류 발생: {e}")

a = 참여자_명단_삽입()
print(a)
```
